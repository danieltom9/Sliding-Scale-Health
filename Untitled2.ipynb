```python
df = pd.read_csv('/Users/danieltom/Desktop/data/DataSet(1).csv')
```

```python
bandwidth_opt = rdd.optimal_bandwidth(df['Sales'], df['Units'], cut=100)
print("Optimal bandwidth:", bandwidth_opt)
```

```python
data_rdd = rdd.truncated_data(df, 'Units', bandwidth_opt, cut=100)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(data_rdd['Units'], data_rdd['Sales'], facecolors='none', edgecolors='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.close()
```

```python




a = data_rdd[data_rdd['Units'] < 100]
aa = np.mean(a['Sales'])
b = data_rdd[data_rdd['Units'] > 100]
bb = np.mean(b['Sales'])
x1 = a['Units']
x2 = b['Units']
x1 = x1.to_numpy()
x2 = x2.to_numpy()
y1 = a['Sales']
y2 = b['Sales']
y1 = y1.to_numpy()
y2 = y2.to_numpy()
m1, b1 = np.polyfit(x1, y1, 1)
plt.plot(x1, y1, 'o')
plt.plot(x1, m1*x1 + b1)
m2, b2 = np.polyfit(x2, y2, 1)
plt.plot(x2, y2, 'o')
plt.plot(x2, m2*x2 + b2)
```
